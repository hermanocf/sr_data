```python
df2 = pd.read_csv('_202211_freq_event_table_test__20221108160521.csv', infer_datetime_format=True)
```

```python
groups=df2['time_difference'].gt('00:00:02').cumsum()
```

```python
df2['CumSum'] = groups
```

```python
df3 = df2['CumSum'].value_counts()
```

```python
df4 = df3.multiply(2)
```

```python
print(df4.median())
print(df4.mean())
print(df4.max())
print(df4.min())
print(df4.mode())
```
